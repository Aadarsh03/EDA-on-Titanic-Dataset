# Summary Statistics 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

## Target Summary

```python
df=pd.read_csv('train.csv')
```

## Load Dataset

```python
df
```

```python
df.tail()
```

```python
print(df.describe())
```

```python
x=df['Survived']
y=df['Pclass']
```

```python
plt.bar(x,y,color='maroon',width=0.4)
plt.xlabel('Survived')
plt.ylabel('Pclass')
plt.title('People Survived in different Classes')
plt.show()
```

```python
# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Countplot of Survived
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df, palette='viridis')
plt.title('Count of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot of Survived with respect to Pclass
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Pclass', data=df, palette='muted')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Pclass')
plt.show()
```

```python
# Boxplot of Fare with respect to Pclass
plt.figure(figsize=(12, 8))
sns.boxplot(x='Pclass', y='Fare', data=df, palette='pastel')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pair plot for numerical features
plt.figure(figsize=(12, 10))
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='husl')
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()
```

```python
# Pie chart for gender distribution
plt.figure(figsize=(8, 8))
gender_distribution = df['Sex'].value_counts()
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
plt.title('Gender Distribution')
plt.show()
```

```python
# Bar chart for SibSp (Number of Siblings/Spouses Aboard)
plt.figure(figsize=(10, 6))
sns.countplot(x='SibSp', data=df, palette='pastel')
plt.title('Number of Siblings/Spouses Aboard')
plt.xlabel('SibSp')
plt.ylabel('Count')
plt.show()
```

```python

# Violin plot for Age distribution with respect to Survived
plt.figure(figsize=(12, 8))
sns.violinplot(x='Survived', y='Age', data=df, palette='Set2')
plt.title('Age Distribution with Respect to Survived')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
# Count of passengers with or without parents/children (Parch) based on Pclass
parch_df = df.groupby(['Pclass', 'Parch']).size().unstack()
parch_df.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 6))
plt.title('Count of Passengers with or without Parents/Children based on Pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.legend(title='Parch')
plt.show()
```

```python
# Comparison of survival rates based on Embarked and Sex
survival_rate_df = df.groupby(['Embarked', 'Sex'])['Survived'].mean().unstack()
survival_rate_df.plot(kind='bar', color=['skyblue', 'salmon'], figsize=(10, 6))
plt.title('Survival Rates based on Embarked and Sex')
plt.xlabel('Embarked')
plt.ylabel('Survival Rate')
plt.legend(title='Sex')
plt.show()
```
